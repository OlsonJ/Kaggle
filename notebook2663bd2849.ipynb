```python
df = pd.read_csv(r'/kaggle/input/datasets/prasertk/top-japanese-game-publishers-stock-price/japanese game publishers stock price 2000-2023.csv')
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# reset index after sort
df = df.sort_values(['Company', 'Date']).reset_index(drop=True)
```

```python
# Verify companies have own row counts
print(df['Company'].value_counts())
```

```python

# Step 5 — Check for duplicate per company
dupes = df.groupby(['Company', 'Date']).size()
print(dupes[dupes > 1])  # Should be empty
```

```python
# use loc to search through index for zero volume
# use np.where for cleaner format : where(col < x, true, false)

df['Z_Volume'] = np.where(df['Volume'] < 1, 1, 0)
```
